````python
#exercice33 - Write a Python function that reads a file and prints the longest line in the file
def plus_longue_ligne(nom_fichier):
    with open(nom_fichier, "r") as fichier:
        lignes = fichier.readlines()
        if lignes:
            plus_longue = max(lignes, key=len)
            print(f"Plus longue ligne: {plus_longue.strip()}")

plus_longue_ligne("document.txt")
````

````python
#exercice34 - Write a Python script to count the frequency of each letter in a text file "message.txt"
with open("message.txt", "r") as fichier:
    contenu = fichier.read()

frequence = {}
for lettre in contenu.lower():
    if lettre.isalpha():
        frequence[lettre] = frequence.get(lettre, 0) + 1

for lettre, compte in sorted(frequence.items()):
    print(f"{lettre}: {compte}")
````

````python
#exercice36 - Create a Python script to check whether a file is empty or not
import os

def fichier_est_vide(nom_fichier):
    taille = os.path.getsize(nom_fichier)
    if taille == 0:
        print("Le fichier est vide")
    else:
        print(f"Le fichier contient {taille} octets")

fichier_est_vide("example.txt")
````

````python
#exercice37 - Write a Python function that reads a file and removes all punctuation marks from the text
import string

def enlever_ponctuation(nom_fichier):
    with open(nom_fichier, "r") as fichier:
        contenu = fichier.read()
    
    sans_ponctuation = ""
    for char in contenu:
        if char not in string.punctuation:
            sans_ponctuation += char
    
    print(sans_ponctuation)

enlever_ponctuation("document.txt")
````

````python
#exercice38 - Create a Python function that writes a list of strings to a file, each string on a new line
def ecrire_liste(liste_strings, nom_fichier):
    with open(nom_fichier, "w") as fichier:
        for ligne in liste_strings:
            fichier.write(ligne + "\n")
    print("Liste écrite dans le fichier!")

ecrire_liste(["Première ligne", "Deuxième ligne", "Troisième ligne"], "output.txt")
````

````python
#exercice39 - Write a Python script to count the number of lines that start with a specific letter (e.g., "A") in a text file
def compter_lignes_lettre(nom_fichier, lettre):
    with open(nom_fichier, "r") as fichier:
        lignes = fichier.readlines()
        compte = 0
        for ligne in lignes:
            if ligne.strip().startswith(lettre):
                compte += 1
        print(f"Nombre de lignes commençant par '{lettre}': {compte}")

compter_lignes_lettre("document.txt", "A")
````

````python
#exercice40 - Write a Python function that takes a string as input and writes it to a file "output.txt" in reverse order
def ecrire_inverse(texte, nom_fichier):
    texte_inverse = texte[::-1]
    with open(nom_fichier, "w") as fichier:
        fichier.write(texte_inverse)
    print("Texte inversé écrit!")

ecrire_inverse("Bonjour monde", "output.txt")
````

````python
#exercice43 - Write a Python script to find the most frequent word in a text file "book.txt"
with open("book.txt", "r") as fichier:
    contenu = fichier.read()

mots = contenu.lower().split()

frequence = {}
for mot in mots:
    frequence[mot] = frequence.get(mot, 0) + 1

mot_plus_frequent = max(frequence, key=frequence.get)
print(f"Mot le plus fréquent: {mot_plus_frequent}")
````

````python
#exercice47 - Create a Python function that opens a file and returns the number of unique words in the file
def mots_uniques(nom_fichier):
    with open(nom_fichier, "r") as fichier:
        contenu = fichier.read()
        mots = set(contenu.split())
        return len(mots)

nombre = mots_uniques("document.txt")
print(f"Nombre de mots uniques: {nombre}")
````

````python
#exercice49 - Write a Python function that reads a file and returns the most common letter in the text
def lettre_plus_frequente(nom_fichier):
    with open(nom_fichier, "r") as fichier:
        contenu = fichier.read()
    
    frequence = {}
    for lettre in contenu.lower():
        if lettre.isalpha():
            frequence[lettre] = frequence.get(lettre, 0) + 1
    
    lettre_max = max(frequence, key=frequence.get)
    print(f"Lettre la plus fréquente: {lettre_max}")

lettre_plus_frequente("document.txt")
````
