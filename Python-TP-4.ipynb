````python
#exercice8
def comparer_nombres(a, b):
    if a == b:
        return "Égaux"
    elif a > b:
        return f"{a} est plus grand"
    else:
        return f"{b} est plus grand"

print(comparer_nombres(5, 10))
````

````python
#exercice10
def sont_anagrammes(str1, str2):
    return sorted(str1.lower()) == sorted(str2.lower())

print(sont_anagrammes("listen", "silent"))
````

````python
#exercice11
def est_premier(nombre):
    if nombre < 2:
        return False
    for i in range(2, nombre):
        if nombre % i == 0:
            return False
    return True

print(est_premier(7))
````

````python
#exercice12 
def est_trie(liste):
    for i in range(len(liste)-1):
        if liste[i] > liste[i+1]:
            return False
    return True

print(est_trie([1, 2, 3, 4, 5]))
````

````python
#exercice23 
def plus_grand_que_reverse(nombre):
    reverse = int(str(nombre)[::-1])
    return nombre > reverse

print(plus_grand_que_reverse(123))
````

````python
#exercice24 
def mot_de_passe_valide(mdp):
    a_lettres = any(c.isalpha() for c in mdp)
    a_chiffres = any(c.isdigit() for c in mdp)
    return a_lettres and a_chiffres

print(mot_de_passe_valide("abc123"))
````

````python
#exercice27
def est_carre_parfait(nombre):
    import math
    racine = math.isqrt(nombre)
    return racine * racine == nombre

print(est_carre_parfait(16))
````

````python
#exercice  48
import math
def est_premier_optimise(nombre):
    if nombre < 2:
        return False
    for i in range(2, int(math.sqrt(nombre)) + 1):
        if nombre % i == 0:
            return False
    return True

print(est_premier_optimise(17))
````
